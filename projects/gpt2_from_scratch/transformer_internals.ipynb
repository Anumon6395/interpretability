### Embedding

Objectives:
1. Learn a matrix that acts as a lookup table to map tokens to their vector representations in the residual stream

tokens of [batch, position] as input. Position is a list of integers. This matrix should be of size d_model * d_vocab.


```python
class Embed(nn.Module):
  def __init__(self, cfg):
    super().__init__()
    self.cfg = cfg
    self.W_E = nn.Parameter(torch.randn(cfg.d_vocab, cfg.d_model, std=cfg.init_range))

  def forward(self, tokens):
    if cfg.debug:
      print(f"tokens: {tokens.shape}")

    embed = self.W_E[tokens, :] # [batch, position, d_model]

    if cfg.debug:
      print(f"W_E: {self.W_E.shape}")

    return embed
```
